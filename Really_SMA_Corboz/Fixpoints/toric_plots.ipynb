```python
import numpy as np
import matplotlib.pyplot as plt

datas = np.load("datas/np_toric_bx0.2bz0.2Dmps4.npy")
p = np.linspace(0, 2.0, 11)
Ceqs = -np.log(datas[2,:])
Cdiffs = -np.log(datas[3,:])
fig, ax1 = plt.subplots(figsize = (6,5),dpi=160)
for i in range(Ceqs.shape[1]):
    plt.plot(p,Ceqs[:,i], color = 'royalblue', lw= 0,marker = 'o', ms = 8, mec= 'royalblue', mfc= 'none', mew = 1.2, label = "Vacuum" if i==0 else "")
for i in range(Cdiffs.shape[1]):
    plt.plot(p,Cdiffs[:,i], color = 'tomato', lw= 0,marker = 'o', ms = 8, mec= 'tomato', mfc= 'none', mew = 1.2, label = "Charge" if i==0 else "")
plt.xlabel('$k_1$/$\pi$')
plt.ylabel('-log(|$\lambda$|)')
plt.ylim(0.5,)
plt.legend()
plt.show()

datas = np.load("datas/np_toric_bx0.2bz0.2Dmps4_domain.npy")
p = np.linspace(0, 2.0, 11)
Ceqs = -np.log(datas[2,:])
Cdiffs = -np.log(datas[3,:])
fig, ax1 = plt.subplots(figsize = (6,5),dpi=160)
for i in range(Ceqs.shape[1]):
    plt.plot(p,Ceqs[:,i], color = 'mediumseagreen', lw= 0,marker = 'o', ms = 8, mec= 'mediumseagreen', mfc= 'none', mew = 1.2, label = "Flux" if i==0 else "")
for i in range(Cdiffs.shape[1]):
    plt.plot(p,Cdiffs[:,i], color ='deeppink', lw= 0,marker = 'o', ms = 8, mec= 'deeppink', mfc= 'none', label = "Fermion" if i==0 else "")
plt.xlabel('$k_1$/$\pi$')
plt.ylabel('-log(|$\lambda$|)')
plt.ylim(2.0,)
plt.legend()
plt.show()
```
